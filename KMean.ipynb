```python
import matplotlib.pyplot as plt
import numpy as np

xyz=np.array(np.random.random((100,3)))
prq=np.array(np.random.random((200,4)))

plt.scatter(xyz[:,0], xyz[:,1])
plt.scatter(prq[:,0], prq[:,1])
plt.show()

```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from numpy import random, float

model = KMeans(n_clusters=2)
model = model.fit(scale(xyz))

model1 = KMeans(n_clusters=2)
model1 = model1.fit(scale(prq))

# We can look at the clusters each data point was assigned to
print(model.labels_) 

plt.figure(figsize=(8, 6))
plt.scatter(prq[:,0], prq[:,1], c=model1.labels_.astype(float))
plt.show()

plt.scatter(xyz[:,0], xyz[:,1], c=model.labels_.astype(float))
plt.show()
```
